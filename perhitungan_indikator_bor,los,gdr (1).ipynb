```python
import pandas as pd

df1 = pd.read_csv('dataset/dataset.csv')
df2 = pd.read_csv('dataset/dataset2.csv')

df = pd.concat([df1, df2], ignore_index=True)
df
```

```python
kolom_dipakai = [
    'no_rawat',
    'tgl_masuk',
    'tgl_keluar',
    'stts_pulang'
]
df = df[kolom_dipakai]
df
```

```python
df['tgl_masuk'] = pd.to_datetime(df['tgl_masuk'], errors='coerce', format='mixed')
df['tgl_keluar'] = pd.to_datetime(df['tgl_keluar'], errors='coerce', format='mixed')
df.info()
```

```python
df['bulan'] = df['tgl_keluar'].dt.month
df['tahun'] = df['tgl_keluar'].dt.year
```

```python
df['tgl_masuk'] = pd.to_datetime(df['tgl_masuk'])
df['tgl_keluar'] = pd.to_datetime(df['tgl_keluar'])

df['bulan_masuk'] = df['tgl_masuk'].dt.month
df['tahun_masuk'] = df['tgl_masuk'].dt.year

df['bulan_keluar'] = df['tgl_keluar'].dt.month
df['tahun_keluar'] = df['tgl_keluar'].dt.year
```

```python
df_masuk = df.groupby(['bulan_masuk', 'tahun_masuk']).agg(
    pasien_masuk=('no_rawat', 'nunique')
).reset_index()
```

```python
df_keluar = df.groupby(['bulan_keluar', 'tahun_keluar']).agg(
    pasien_keluar=('no_rawat', 'nunique'),
    pasien_meninggal=('stts_pulang', lambda x: (x.str.lower() == 'meninggal').sum()),
    total_hari_perawatan=(
        'tgl_keluar',
        lambda x: (x - df.loc[x.index, 'tgl_masuk']).dt.days.sum()
    ),
    tempat_tidur=('jumlah_bed', 'first')
).reset_index()
```

penggabungan data dilakukan menggunakan operasi merge berdasarkan kesamaan bulan dan tahun

```python
df_rekap = pd.merge(
    df_keluar,
    df_masuk,
    left_on=['bulan_keluar', 'tahun_keluar'],
    right_on=['bulan_masuk', 'tahun_masuk'],
    how='left'
)
```

```python
df_rekap = df_rekap.rename(columns={
    'bulan_keluar': 'bulan',
    'tahun_keluar': 'tahun'
})

df_rekap = df_rekap[
    ['bulan', 'tahun', 'pasien_masuk', 'pasien_keluar',
     'pasien_meninggal', 'total_hari_perawatan', 'tempat_tidur']
]
```

Nilai periode_bulan dihitung sebagai jumlah hari aktual pada setiap bulan berdasarkan tahun

```python
df_rekap['periode_bulan'] = df_rekap.apply(
    lambda x: calendar.monthrange(int(x['tahun']), int(x['bulan']))[1],
    axis=1
)
```

```python
df_rekap['BOR'] = (
    df_rekap['total_hari_perawatan'] /
    (df_rekap['tempat_tidur'] * df_rekap['periode_bulan'])
) * 100
```

```python
df_rekap['LOS'] = (
    df_rekap['total_hari_perawatan'] /
    df_rekap['pasien_keluar']
)
```

```python
df_rekap['GDR'] = (
    df_rekap['pasien_meninggal'] /
    df_rekap['pasien_keluar']
) * 1000
```

```python
df_rekap = df_rekap.sort_values(by=['tahun', 'bulan']).reset_index(drop=True)
df_rekap
```

```python
df_indikator = df_rekap[['bulan', 'tahun', 'BOR', 'LOS', 'GDR']]
df_indikator
```

```python
rata_bor = df_rekap['BOR'].mean()
rata_los = df_rekap['LOS'].mean()
```

```python
gdr_tahunan = (
    df_rekap['pasien_meninggal'].sum() /
    df_rekap['pasien_keluar'].sum()
) * 1000
```

```python
df_ringkas_tahun = pd.DataFrame({
    'tahun': [2025],
    'BOR_rata_rata': [round(rata_bor, 2)],
    'LOS_rata_rata': [round(rata_los, 2)],
    'GDR_tahunan_per_1000': [round(gdr_tahunan, 2)]
})

df_ringkas_tahun
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_scaled = df_rekap.copy()

df_scaled[fitur] = scaler.fit_transform(df_rekap[fitur])
```

```python
import numpy as np

def euclidean_distance(case1, case2):
    return np.sqrt(np.sum((case1 - case2)**2))
```

```python
# contoh kasus baru
kasus_baru = np.array([[0.60, 0.50, 0.30]])  # nanti pakai scaler juga

kasus_baru_scaled = scaler.transform([[60, 4.5, 10]])
```

```python
distances = []

for i in range(len(df_scaled)):
    case_lama = df_scaled.loc[i, fitur].values
    jarak = euclidean_distance(kasus_baru_scaled[0], case_lama)
    distances.append(jarak)

df_scaled['Distance'] = distances
```

```python
hasil = df_scaled.sort_values('Distance').iloc[0]
hasil
```
