```python
df_masuk = df.groupby(['bulan_masuk', 'tahun_masuk']).agg(
    pasien_masuk=('no_rawat', 'nunique')
).reset_index()
```

```python
df_keluar = df.groupby(['bulan_keluar', 'tahun_keluar']).agg(
    pasien_keluar=('no_rawat', 'nunique'),
    pasien_meninggal=('stts_pulang', lambda x: (x.str.lower() == 'meninggal').sum()),
    total_hari_perawatan=(
        'tgl_keluar',
        lambda x: (x - df.loc[x.index, 'tgl_masuk']).dt.days.sum()
    ),
    tempat_tidur=('jumlah_bed', 'first')
).reset_index()
```

penggabungan data dilakukan menggunakan operasi merge berdasarkan kesamaan bulan dan tahun

```python
df_rekap = pd.merge(
    df_keluar,
    df_masuk,
    left_on=['bulan_keluar', 'tahun_keluar'],
    right_on=['bulan_masuk', 'tahun_masuk'],
    how='left'
)
```

```python
df_rekap = df_rekap.rename(columns={
    'bulan_keluar': 'bulan',
    'tahun_keluar': 'tahun'
})

df_rekap = df_rekap[
    ['bulan', 'tahun', 'pasien_masuk', 'pasien_keluar',
     'pasien_meninggal', 'total_hari_perawatan', 'tempat_tidur']
]
```

Nilai periode_bulan dihitung sebagai jumlah hari aktual pada setiap bulan berdasarkan tahun

```python
df_rekap['periode_bulan'] = df_rekap.apply(
    lambda x: calendar.monthrange(int(x['tahun']), int(x['bulan']))[1],
    axis=1
)
```

```python
df_rekap['BOR'] = (
    df_rekap['total_hari_perawatan'] /
    (df_rekap['tempat_tidur'] * df_rekap['periode_bulan'])
) * 100
```

```python
df_rekap['LOS'] = (
    df_rekap['total_hari_perawatan'] /
    df_rekap['pasien_keluar']
)
```

```python
df_rekap['GDR'] = (
    df_rekap['pasien_meninggal'] /
    df_rekap['pasien_keluar']
) * 1000
```

```python
df_rekap = df_rekap.sort_values(by=['tahun', 'bulan']).reset_index(drop=True)
df_rekap
```

```python
from sklearn.model_selection import train_test_split
from sklearn.multioutput import MultiOutputClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import hamming_loss
```

```python
X = df_merge[['BOR', 'LOS', 'GDR']]
label_cols = df_merge.columns[5:]
y = df_merge[label_cols].astype(int)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b)**2))
```

```python
def cbr_predict(X_train, y_train, X_test):
    predictions = []

    for i in range(len(X_test)):
        test_case = X_test.iloc[i].values
        
        distances = []
        
        # bandingkan ke semua data training
        for j in range(len(X_train)):
            train_case = X_train.iloc[j].values
            dist = euclidean_distance(test_case, train_case)
            distances.append(dist)
        
        # ambil index paling mirip
        nearest_index = np.argmin(distances)
        
        # ambil solusi (label)
        pred_label = y_train.iloc[nearest_index].values
        predictions.append(pred_label)
    
    return np.array(predictions)
```

```python
y_pred = cbr_predict(X_train, y_train, X_test)
```

```python
hl = hamming_loss(y_test, y_pred)
print("Hamming Loss:", hl)
```

```python
df_merge.columns = df_merge.columns.str.strip()

print(df_merge.columns.tolist())
```

```python
df_merge.columns = df_merge.columns.str.strip()

fitur_cols = ['BOR', 'LOS', 'GDR']

label_cols = [col for col in df_merge.columns 
              if col not in ['bulan', 'tahun', 'periode'] + fitur_cols]
X = df_merge[fitur_cols].values
y = df_merge[label_cols].astype(int).values
```

```python
df_merge['periode'] = df_merge['tahun'].astype(str) + '-' + df_merge['bulan'].astype(str).str.zfill(2)
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# normalisasi data
scaler = MinMaxScaler()
X = scaler.fit_transform(X)

def cbr_predict(X_train, y_train, X_test, threshold=0.5):
    y_pred = []

    for x in X_test:
        # hitung jarak Euclidean
        distances = np.linalg.norm(X_train - x, axis=1)

        # ambil index dengan jarak <= threshold
        idx = np.where(distances <= threshold)[0]

        if len(idx) == 0:
            # kalau tidak ada yang mirip
            y_pred.append(np.zeros(y_train.shape[1]))
        else:
            # ambil label dari kasus yang mirip
            selected_labels = y_train[idx]

            # voting (rata-rata)
            pred = (selected_labels.mean(axis=0) >= 0.5).astype(int)

            y_pred.append(pred)

    return np.array(y_pred)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import hamming_loss

# threshold
thresholds = np.arange(0.4, 0.61, 0.05)

results = []

# ambil urutan periode
periodes = sorted(df_merge['periode'].unique())

# loop mulai dari periode ke-2 (biar ada data train)
for i in range(1, len(periodes)):
    
    # split data
    train = df_merge[df_merge['periode'] < periodes[i]]
    test  = df_merge[df_merge['periode'] == periodes[i]]
    
    if train.empty or test.empty:
        continue
    
    # fitur & label
    X_train = train[fitur_cols].values
    y_train = train[label_cols].astype(int).values
    
    X_test = test[fitur_cols].values
    y_test = test[label_cols].astype(int).values
    
    # scaling
    scaler = MinMaxScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    
    # loop threshold
    for t in thresholds:
        
        # prediksi
        y_pred = cbr_predict(X_train, y_train, X_test, threshold=t)
        
        # evaluasi
        loss = hamming_loss(y_test, y_pred)
        
        coverage = np.sum(np.sum(y_pred, axis=1) > 0) / len(y_pred)
        
        # simpan hasil
        results.append({
            'periode': periodes[i],
            'threshold': t,
            'loss': loss,
            'coverage': coverage
        })
        
        print(f"{periodes[i]} | T={t} | Loss={loss:.4f} | Cov={coverage:.2f}")

```

```python
df_results = pd.DataFrame(results)
summary = df_results.groupby('threshold').agg({
    'loss': 'mean',
    'coverage': 'mean'
}).reset_index()

summary
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from openpyxl import load_workbook
from openpyxl.drawing.image import Image

# simpan tabel ke excel
summary.to_excel('hasil_threshold.xlsx', index=False)

# posisi batang
x = np.arange(len(summary['threshold']))
width = 0.35

# buat figure
plt.figure(figsize=(8,5))

# batang loss
plt.bar(
    x - width/2,
    summary['loss'],
    width,
    label='Hamming Loss'
)

# batang coverage
plt.bar(
    x + width/2,
    summary['coverage'],
    width,
    label='Coverage'
)

# label sumbu x
plt.xticks(x, summary['threshold'])

# judul & label
plt.xlabel('Threshold')
plt.ylabel('Nilai')
plt.title('Perbandingan Threshold terhadap Hamming Loss dan Coverage')

# legenda
plt.legend()

# grid
plt.grid(True, axis='y')

# simpan grafik
plt.savefig('grafik_batang_threshold.png')
plt.close()

# buka excel
wb = load_workbook('hasil_threshold.xlsx')
ws = wb.active

# masukkan gambar ke excel
img = Image('grafik_batang_threshold.png')
ws.add_image(img, 'F2')

# simpan file baru
wb.save('hasil_threshold_dengan_grafik.xlsx')
```
